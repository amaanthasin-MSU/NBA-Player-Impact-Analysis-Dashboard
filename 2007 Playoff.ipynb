```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
player_stats_data = pd.read_csv('player_stats_ppg.csv')
team_records_data = pd.read_csv('records.csv')

# Clean player_stats_data
player_stats_data = player_stats_data.dropna(subset=['Player'])  # Dropping the initial NaN row
player_stats_data = player_stats_data[['Player', 'Tm', 'PTS', 'TRB', 'AST']]  # Keeping relevant statistics
player_stats_data['PTS'] = pd.to_numeric(player_stats_data['PTS'], errors='coerce')
player_stats_data['TRB'] = pd.to_numeric(player_stats_data['TRB'], errors='coerce')
player_stats_data['AST'] = pd.to_numeric(player_stats_data['AST'], errors='coerce')

# Clean team_records_data
team_records_data = team_records_data[['Team', 'Overall']]  # Keeping only relevant columns
team_records_data['Wins'] = team_records_data['Overall'].str.split('-').str[0].astype(int)
team_records_data['Losses'] = team_records_data['Overall'].str.split('-').str[1].astype(int)
team_records_data['Win Percentage'] = team_records_data['Wins'] / (team_records_data['Wins'] + team_records_data['Losses'])

# Merge the cleaned data
merged_stats_records = pd.merge(player_stats_data, team_records_data, left_on='Tm', right_on='Team')

plt.figure(figsize=(12, 6))
sns.regplot(x='PTS', y='Win Percentage', data=merged_stats_records, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})

plt.title('Scatter Plot with Regression Line: Points Scored vs. Team Win Percentage')
plt.xlabel('Points Scored')
plt.ylabel('Team Win Percentage')
plt.grid(True)
plt.show()
```

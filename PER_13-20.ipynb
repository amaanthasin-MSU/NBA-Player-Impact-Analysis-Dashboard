```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
champ_20 = pd.read_csv("champ_20.csv")
champ_19 = pd.read_csv("champ_19.csv")
champ_18 = pd.read_csv("champ_18.csv")
champ_17 = pd.read_csv("champ_17.csv")
champ_16 = pd.read_csv("champ_16.csv")
champ_15 = pd.read_csv("champ_15.csv")
champ_14 = pd.read_csv("champ_14.csv")
champ_13 = pd.read_csv("champ_13.csv")

worst_20 = pd.read_csv("worst_20.csv")
worst_19 = pd.read_csv("worst_19.csv")
worst_18 = pd.read_csv("worst_18.csv")
worst_17 = pd.read_csv("worst_17.csv")
worst_16 = pd.read_csv("worst_16.csv")
worst_15 = pd.read_csv("worst_15.csv")
worst_14 = pd.read_csv("worst_14.csv")
worst_13 = pd.read_csv("worst_13.csv")

champ_14
```

```python
filtered_df = champ_20[champ_20['G'] >= 60]
top_2_per = filtered_df.sort_values(by='PER', ascending=False).head(2)
avg_top_20 = top_2_per["PER"].mean()
```

```python
def calculate_avg_per(df):
    filtered_df = df[df['G'] >= 60]
    top_2_per = filtered_df.sort_values(by='PER', ascending=False).head(2)
    worst_2_per = filtered_df.sort_values(by='PER', ascending=False).head(2)
    return top_2_per['PER'].mean() 


avg_per_20 = calculate_avg_per(champ_20)
avg_per_19 = calculate_avg_per(champ_19)
avg_per_18 = calculate_avg_per(champ_18)
avg_per_17 = calculate_avg_per(champ_17)
avg_per_16 = calculate_avg_per(champ_16)
avg_per_15 = calculate_avg_per(champ_15)
avg_per_14 = calculate_avg_per(champ_14)
avg_per_13 = calculate_avg_per(champ_13)

print("Average PER 2020:", avg_per_20)
print("Average PER 2019:", avg_per_19)
print("Average PER 2018:", avg_per_18)
print("Average PER 2017:", avg_per_17)
print("Average PER 2016:", avg_per_16)
print("Average PER 2015:", avg_per_15)
print("Average PER 2014:", avg_per_14)
print("Average PER 2013:", avg_per_13)
```

```python
def calculate_avg_worst_per(df):
    filtered_df = df[df['G'] >= 60]
    worst_2_per = filtered_df.sort_values(by='PER', ascending=False).head(2)
    return worst_2_per['PER'].mean()


avg_worst_per_20 = calculate_avg_worst_per(worst_20)
avg_worst_per_19 = round(calculate_avg_worst_per(worst_19),2)
avg_worst_per_18 = calculate_avg_worst_per(worst_18)
avg_worst_per_17 = round(calculate_avg_worst_per(worst_17),2)
avg_worst_per_16 = calculate_avg_worst_per(worst_16)
avg_worst_per_15 = round(calculate_avg_worst_per(worst_15),2)
avg_worst_per_14 = calculate_avg_worst_per(worst_14)
avg_worst_per_13 = calculate_avg_worst_per(worst_13)

print("Average Worst PER 2020:", avg_worst_per_20)
print("Average Worst PER 2019:", round(avg_worst_per_19),2)
print("Average Worst PER 2018:", avg_worst_per_18)
print("Average Worst PER 2017:", round(avg_worst_per_17),2)
print("Average Worst PER 2016:", avg_worst_per_16)
print("Average Worst PER 2015:", round(avg_worst_per_15),2)
print("Average Worst PER 2014:", avg_worst_per_14)
print("Average Worst PER 2013:", avg_worst_per_13)
```

```python
import matplotlib.pyplot as plt

# Years and PER data
years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
avg_best_per = [avg_per_13, avg_per_14, avg_per_15, avg_per_16, avg_per_17, avg_per_18, avg_per_19, avg_per_20]
avg_worst_per = [avg_worst_per_13, avg_worst_per_14, avg_worst_per_15, avg_worst_per_16, avg_worst_per_17, avg_worst_per_18, avg_worst_per_19, avg_worst_per_20]

years_int = [int(year) for year in years]

fig, ax = plt.subplots(figsize=(10, 6))

ax.fill_between(years_int, avg_best_per, color="blue", alpha=0.4, label='Best Team Top 2 PER')
ax.fill_between(years_int, avg_worst_per, color="pink", alpha=0.5, label='Worst Team Top 2 PER')

ax.set_xlabel('Year')
ax.set_ylabel('Average PER')
ax.set_title('Average PER of Top 2 Players on Best vs. Worst NBA Teams (2013-2020)')
plt.xticks(years_int, years)  
ax.legend()
plt.savefig('PER_13-20', dpi=120)
plt.show()
```
